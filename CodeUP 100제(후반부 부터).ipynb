```python
# 6093
n = int(input())
a = input().split()
stack = []
for i in range(n):
    a[i] = int(a[i])
    stack.append(a[i])
    
for j in range(n):
    print(stack.pop(), end=" ")
```

```python
# 6094
n = int(input())
a = input().split()
min = int(a[0])

for i in range(1, n):
    a[i] = int(a[i])
    if a[i]<=min: min=a[i]
    
print(min)
```

```python
# 6095
p = []
for i in range(19):
    p.append([]) # 리스트 안에 다른 리스트 추가하기
    for j in range(19):
        p[i].append(0)
        
n = int(input())
for j in range(n):
    x, y = map(int, input().split())
    p[x-1][y-1]=1
    
for i in range(19):
    for j in range(19):
        print(p[i][j], end=' ')
    print()
```
